```python
X_train, X_test= train_test_split(df, train_size=0.8, random_state=50)
```

```python
#Label Encoding
from keras.utils.np_utils import to_categorical    #one hot encoding yapar kerasın anlayacağı şekil

y_train = to_categorical(target_train, num_classes=3)
y_test = to_categorical(target_test, num_classes=3)
```

```python
train_dataset = train_dataset.drop(columns = ["Timestamp", "Protocol","PSH_Flag_Cnt","Init_Fwd_Win_Byts","Flow_Byts/s","Flow_Pkts/s", "Label"], axis=1)
test_dataset = test_dataset.drop(columns = ["Timestamp", "Protocol","PSH_Flag_Cnt","Init_Fwd_Win_Byts","Flow_Byts/s","Flow_Pkts/s", "Label"], axis=1)

# axis=1 -> sütunların temsili
# protocol          -> bütün değerler 6
#PSH_Flag_Cnt       -> PSHli paket sayısı
#Init_Fwd_Win_Byts  -> İlk pencerede ileri yönde gönderilen bayt sayısı
#Flow_Byts/s        -> Saniyede aktarılan paket sayısı olan akış bayt oranı 
#Flow_Pkts/s        -> Saniyede aktarılan paket sayısı olan akış paket hızı
```

```python
X_train = train_dataset.iloc[:, :-1].values
X_test = test_dataset.iloc[:, :-1].values
X_test

#iloc[:, :-1].values -> son sütunun değerlerinin satır vektörünü verir.
```

```python
#boyutları
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

CNN için verileri yeniden şekillendirin

```python
X_train = X_train.reshape(len(X_train), X_train.shape[1], 1)
X_test = X_test.reshape(len(X_test), X_test.shape[1], 1)
X_train.shape, X_test.shape

#Boyutları aynı olmalı ve ,,1 olmalı 
```

```python
#derin öğrenme işlevini yapmak
#relu ->activation fonksiyonu
from keras.models import Sequential         #modelimizi import ediyoruz.
from keras.layers import Conv2D, Conv1D, MaxPooling2D, MaxPooling1D, Flatten, BatchNormalization, Dense

def model():
    model = Sequential()   #Sequential -> Bütün Layerlara verilen ad.
    
    model.add(Conv1D(filters=64, kernel_size=6, activation='relu',       #kernel_size -> filtre boyutu 
                    padding='same', input_shape=(72, 1)))
    model.add(BatchNormalization())
    # havuz katmanı ekleme
    model.add(MaxPooling1D(pool_size=(3), strides=2, padding='same'))
    
    
    model.add(Conv1D(filters=64, kernel_size=6, activation='relu', 
                    padding='same', input_shape=(72, 1)))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size=(3), strides=2, padding='same'))
    
    
    model.add(Conv1D(filters=64, kernel_size=6, activation='relu', 
                    padding='same', input_shape=(72, 1)))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size=(3), strides=2, padding='same'))
    
    model.add(Flatten())    #düzleştirme işlemi
    model.add(Dense(64, activation='relu'))      # 64 tane nöron
    model.add(Dense(64, activation='relu'))
    model.add(Dense(3, activation='softmax'))    #  output   softmax -> sonuçları yüzdelik dilimlere ayırır   dense(3) -> class sayısı kadar
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) #loss kullanmamızın sebebi softmax kullanmış olmamız
    return model
```

```python
#Keras Modelini ve Özetini Derleme

model = model()
model.summary()
```

```python
from keras.callbacks import CSVLogger, ModelCheckpoint

logger = CSVLogger('logs.csv', append=True)
his = model.fit(X_train, y_train, epochs=30, batch_size=32,           
          validation_data=(X_test, y_test), callbacks=[logger])

#epoch -> döngü
#batch_size=32 -> her iterasyonda(epochta) 32 tane değeri train ediyor
```

Sonuçların Görselleştirilmesi (CNN) 

Verilerimize CNN uygulanarak elde edilen sonuçların grafiksel bir görüntülenmesini yapalım.

```python
# Test verilerindeki model performansını kontrol edin
scores = model.evaluate(X_test, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1] * 100))
```

```python
# Model geçmişini kontrol edin
#epochlarda gördüğümüz değerleri ekrana yazdırır.

history = his.history
history.keys()


#loss ve accuary          -> train seti üzerinden elde edilen sonuçlar
#val_loss ve val_accuracy -> test seti üzerinden elde edilen sonuçlar
```

```python
epochs = range(1, len(history['loss']) + 1)
acc = history['accuracy']
loss = history['loss']
val_acc = history['val_accuracy']
val_loss = history['val_loss']

#training and val accuracy doğruluğunu görselleştirin.
plt.figure(figsize=(10, 5))
plt.title('Training and Validation Accuracy (CNN)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.plot(epochs, acc, label='accuracy')
plt.plot(epochs, val_acc, label='val_acc')
plt.legend()

#train and val loss görselleştirme
plt.figure(figsize=(10, 5))
plt.title('Training and Validation Loss(CNN)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.plot(epochs, loss, label='loss', color='g')
plt.plot(epochs, val_loss, label='val_loss', color='r')
plt.legend()



#100 epoch ile test edip kaydedecektim ama %100 oldu zaten.

#loss ve accuary          -> train seti üzerinden elde edilen sonuçlar
#val_loss ve val_accuracy -> test seti üzerinden elde edilen sonuçlar
```
